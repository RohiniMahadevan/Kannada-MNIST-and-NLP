#Importing the necessary libraries

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import pytesseract
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

#loading the file through numpy

```python
X_train = np.load(r"C:\Users\rohin\OneDrive\Documents\Kannada_MNIST_datataset_paper\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\X_kannada_MNIST_train.npz")
y_train = np.load(r"C:\Users\rohin\OneDrive\Documents\Kannada_MNIST_datataset_paper\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\y_kannada_MNIST_train.npz")
X_test = np.load(r"C:\Users\rohin\OneDrive\Documents\Kannada_MNIST_datataset_paper\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\X_kannada_MNIST_test.npz")
y_test = np.load(r"C:\Users\rohin\OneDrive\Documents\Kannada_MNIST_datataset_paper\Kannada_MNIST_datataset_paper\Kannada_MNIST_npz\Kannada_MNIST\y_kannada_MNIST_test.npz")
```

#shaping the train and test data array

```python
X_train_data = X_train['arr_0'].reshape(X_train['arr_0'].shape[0], -1)
X_test_data = X_test['arr_0'].reshape(X_test['arr_0'].shape[0], -1)
```

```python
pca = PCA(n_components=10)
```

#fitting and transforming the pca component in the training and test data

```python
X_train_pca = pca.fit_transform(X_train_data)
X_train_pca = X_train_pca.astype(np.float32)
```

```python
X_test_pca = pca.transform(X_test_data)
X_test_pca = X_test_pca.astype(np.float32)

y_train_data = y_train['arr_0']
y_test_data = y_test['arr_0']
```

```python
pca_components = [10, 15, 20, 25, 30]
```

#model = decision tree

```python
for size in pca_components:
    # Perform PCA with the current component size
    pca = PCA(n_components=size)
    X_train_pca = pca.fit_transform(X_train_data)
    X_test_pca = pca.transform(X_test_data)

    # Train the Decision Tree classifier
    dt_classifier = DecisionTreeClassifier()
    dt_classifier.fit(X_train_pca, y_train_data)

    # Predict labels for the test set
    dt_y_pred = dt_classifier.predict(X_test_pca)

    # Calculate precision, recall, F1-score
    classification_metrics = classification_report(y_test_data, dt_y_pred)
    print("Classification Report (Component Size =", size, "):", classification_metrics)

    # Calculate confusion matrix
    cm = confusion_matrix(y_test_data, dt_y_pred)
    print("Confusion Matrix (Component Size =", size, "):\n", cm)

    # Convert y_test_data to binary representation
    y_test_binary = label_binarize(y_test_data, classes=range(10))

    # Calculate predicted probabilities for each class
    y_scores = dt_classifier.predict_proba(X_test_pca)

    # Calculate ROC-AUC score
    auc_score = roc_auc_score(y_test_binary, y_scores, average='macro')
    print("ROC-AUC Score (Component Size =", size, "):", auc_score)

    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_test_binary.ravel(), y_scores.ravel())

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (Component Size = %d)' % size)
    plt.legend(loc="lower right")
    plt.show()
```
